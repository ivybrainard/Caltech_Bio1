```python
def getGenerationWMutations(generations, start, num_alleles, rate):
  frequency= []
  frequency.append(start/num_alleles); ###
  for i in range(1, generations+1):
    flips = np.random.rand(num_alleles)#* (start *2)
    numRed = np.sum(flips < frequency[i-1])

    mutationRedRandom = np.random.rand(numRed);
    mutationWhiteRandom = np.random.rand(num_alleles - numRed);

    redMutataion = np.sum(mutationRedRandom < (rate));
    whiteMutataion = np.sum(mutationWhiteRandom < (rate))

    numRed = numRed - redMutataion + whiteMutataion

    frequency.append(numRed/num_alleles)
  return frequency;


rate = .001;
generations = 109;
flies = 16;
num_alleles = flies * 2;

#plottinh the allele frequency for a single population of 109 generations
plt.plot(getGenerationWMutations(generations, flies, num_alleles, rate))
plt.xlabel("Generations", color = "purple")
plt.title("Allele Frequency for a single population: mutation rate of 0.001", color = "purple")
plt.ylabel("Allele Frequency", color = "purple")
```

```python
#get p and q

def get_p(time, s):
  p = 1/(1+(10**9) * np.exp(-np.log(2)*s*time))
  return p;

def get_q(time, s):
  q = 1 - (get_p(time,s));
  return q;

years = np.linspace(0,100);
hours = years * 8760

```

```python
#s = 10^-4

s = 10**-4


p = get_p(hours, s);
q = get_q(hours, s);

plt.plot(years, p, label="p(t)", color = "pink");
plt.plot(years, q, label = "g(t)", color = "skyblue");
plt.legend()
plt.xlabel("year", color = "purple");
plt.ylabel("Frequency", color = "purple")
plt.title("p(t) and g(t) when s = 10^-4", color = "purple");


```

```python
#s = 10^-3
s = 10**-3

p = get_p(hours, s);
q = get_q(hours, s);

plt.plot(years, p, label="p(t)", color = "pink");
plt.plot(years, q, label = "g(t)", color = "skyblue");
plt.legend()
plt.xlabel("year", color = "purple");
plt.ylabel("Frequency", color = "purple")
plt.title("p(t) and g(t) when s = 10^-3", color = "purple");
```
